```python
import pandas as pd
import numpy as np
```

```python
def add_parent_num(df, parent):
    df['Mutation ID'] = parent + df['Mutation ID']
    return df
```

```python
columns_of_interest = ['Mutation ID', 'Chr', 'Start', 'Ref', 'Alt', 'Otherinfo11', 'Otherinfo12', 'Otherinfo13']
master_file = "CDDR_MasterMaf_Jawad.xlsx"

df1 = pd.read_excel(master_file, sheet_name=0, usecols=columns_of_interest)
df2 = pd.read_excel(master_file, sheet_name=1, usecols=columns_of_interest)
df3 = pd.read_excel(master_file, sheet_name=2, usecols=columns_of_interest)

df1 = add_parent_num(df1, "PCI13")
df2 = add_parent_num(df2, "HN31")
df3 = add_parent_num(df3, "HN30")

mf = pd.concat([df1,df2,df3])
# mf.drop_duplicates(subset='Mutation ID', inplace=True)
mf.reset_index(drop=True, inplace=True)
```

```python
display(mf)
mf.drop_duplicates(subset='Mutation ID', inplace=True)
```

```python
mf.reset_index(drop=True, inplace=True)
display(mf)
```

```python
fdf = pd.read_excel("Final mutations of interest_Jawad.xlsx", usecols=["Parental", "Mutation ID"])
fdf['Mutation ID'] = fdf['Parental'] + fdf['Mutation ID']
```

```python
# removes filtered rows from final output
master_ids = mf['Mutation ID'].to_numpy()
ref_ids = fdf['Mutation ID'].to_numpy()
mf['Remove'] = np.isin(master_ids, ref_ids)
mf = mf.loc[mf.Remove, :]
```

```python
# sorts it based on final output
mf = mf.set_index('Mutation ID')
mf = mf.reindex(index=fdf['Mutation ID'])
mf = mf.reset_index()
# fdf dataframe  2
```
