```python
import pandas as pd
```

```python
df = pd.read_excel("output.xlsx", usecols=[0,1,2,3,4])
```

```python
df.sort_values(by=['SAMPLE_ID'])
```

```python
counts = df.SAMPLE_ID.value_counts()
counts_idxs = list(counts.index)
counts_vals = list(counts.to_numpy())
counts_dict = dict(zip(counts_idxs, counts_vals))
ordered_counts_dict = dict(sorted(counts_dict.items()))
counts_idxs = list(ordered_counts_dict.keys())
counts_vals = list(ordered_counts_dict.values())
vals_idx = []
total = 0
for count, val in enumerate(counts_vals):
    total += val
    vals_idx.append(total)
vals_idx.insert(0, 0)
print(vals_idx)
```

```python
for count in range(len(vals_idx) - 1):
    df2 = df.loc[vals_idx[count]: vals_idx[count+1]-1]
    df2 = df2.drop('SAMPLE_ID', 1)
    df2.insert(2, 'FILTER', 'PASS')
    display(df2)
    df2.to_csv(str(counts_idxs[count]) + ".vcf", index=False, sep='\t')
```
